# Step Two

```python
#Create a for-loop that appends 'days' based on values in 'order_day_of_week'
days = []

for day in df_ords_merged['Order_day_of_week']:
    if day == 0 or day ==1:
        days.append('Busiest days')
    elif day == 4 or day == 3:
        days.append('Slowest days')
    else:
        days.append('Regularly busy days')
```

```python
#create busiest days column
df_ords_merged['Busiest_days'] = days
```

# Step Four

```python
#frequency of 'order_hour_of_day' column
df_ords_merged['order_hour_of_day'].value_counts(dropna = False)
```

Now that we have the frequency of the order hour of days we can break this down into categories.
Most orders (9, 10, 11, 12, 13, 14, 15, 16) 9 am to 4 pm
Average orders (7, 8, 17, 18, 19 ,20, 21, 22) 7-8 am & 5- 10 pm
Fewest orders (0, 1, 2, 3, 4, 5, 6, 23) 11 pm - 6 am


```python
# for loop for busiest hours of day based on 'order_hour_of_day' column
hours = []

for hour in df_ords_merged['order_hour_of_day']:
  if hour in [9, 10, 11, 12, 13, 14, 15, 16]:
    hours.append('Most orders')
  elif hour in [0, 1, 2, 3, 4, 5, 6, 23]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
#create busiest_period_of_day column
df_ords_merged['busiest_period_of_day'] = hours
```
